```python
# Sample healthcare claims dataset
# Demonstrates use of lists and tuples in Python
# Includes one missing CPT code for data quality practice

claims = [
    ("CLMID001", "2026-04-01", "91199", 145.99), 
    ("CLMID002", "2026-04-02", "99223", 34.99), 
    ("CLMID003", "2026-04-03", "99343", 44.99),
    ("CLMID004", "2026-04-04", "99556", 55.99),
    ("CLMID005", "2026-04-05", None, 220.99), 
] 
```

```python
# Find claims over $200
print("High-Cost Claims:")
for claim in claims:
    if claim[3] > 200:
        print(claim)

# Find claims with missing CPT codes
print("\nClaims Missing CPT Codes:")
for claim in claims:
    if claim[2] is None:
        print(claim)
```
